Holdout first

```python
coco_path = Path(
    "/home/pau/Documents/master/M6/project/data/AICity_data/AICity_data/train"
    "/S03/c010/gt_coco"
)
```

```python
image_ids = [x["id"] for x in gt_ann["images"]]
image_ids = list(set(image_ids))
image_ids.sort()

image_ids
```

```python
def create_partition(ids: list, gt: dict):
    images = [x for x in gt["images"] if x["id"] in ids]
    annots = [x for x in gt["annotations"] if x["image_id"] in ids]

    partition = {
        "info": gt["info"],
        "images": images,
        "categories": gt["categories"],
        "annotations": annots,
        "licenses": gt["licenses"]
    }
    return partition
```

```python
thresh = .25
ind = int(len(image_ids) * thresh)

train_ids = image_ids[:ind]
test_ids = image_ids[ind:]

holdout_train = create_partition(train_ids, gt_ann)
holdout_test = create_partition(test_ids, gt_ann)

with open(coco_path / "holdout" / "gt_all_train.json", 'w') as f_json:
    json.dump(holdout_train, f_json)

with open(coco_path / "holdout" / "gt_all_test.json", 'w') as f_json:
    json.dump(holdout_test, f_json)
```

```python
from copy import deepcopy

change = {
    1: 3,
    2: 2,
}

coco_ann = deepcopy(holdout_test)
for ii in coco_ann["categories"]:
    ii["id"] = change[ii["id"]]
for ii in coco_ann["annotations"]:
    ii["category_id"] = change[ii["category_id"]]

with open(coco_path / "gt_all_test_cocoid.json", 'w') as f_json:
    json.dump(coco_ann, f_json)
```

Kfold now

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=4, shuffle=False)
splits = kf.split(image_ids)

for num, (test_ids, train_ids) in enumerate(splits):
    train_ids = train_ids.tolist()
    test_ids = test_ids.tolist()

    kfold_train = create_partition(train_ids, gt_ann)
    kfold_test = create_partition(test_ids, gt_ann)

    (coco_path / f"kfold_{num}").mkdir(exist_ok=True, parents=True)

    with open(coco_path / f"kfold_{num}" / "gt_all_train.json", 'w') as f_json:
        json.dump(kfold_train, f_json)

    with open(coco_path / f"kfold_{num}" / "gt_all_test.json", 'w') as f_json:
        json.dump(kfold_test, f_json)
```
