```python

import matplotlib.pyplot as plt

confidence = [0,0.5,0.8,0.9,0.99,0.999,0.9999]
idf1_ = [0.054131, 0.200617, 0.227923, 0.238627, 0.253072, 0.289069, 0.405224]
plt.figure(dpi=150, figsize=(10, 7))
plt.style.use('ggplot')
plt.title("IDF1 evolution over accepted confident bbox (overfitted)")
plt.xlabel("Accepted object confidence")

ax1 = plt.subplot()
ax1.set_ylim([0, 0.8])
ax1.grid(False)

l1 = ax1.plot(confidence, idf1_, color='c')
ax1.set_ylabel("IDF1")
# ax1.set_xscale('log')
l3 = ax1.axhline(y=idf1_[-1], color='red', linestyle='--', label="IDF1 = 0.41")
plt.legend([l1,l3], ["IDF1","IDF1 = 0.41"])
plt.tight_layout()
plt.savefig("./plots/t21_idf1_OF.png")
plt.show()
plt.close()
```
